```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Case 1
exercise = np.array([0.0, 0.53, 1.05, 1.58, 2.11, 2.63, 3.16, 3.68, 4.21, 4.74, 5.26, 5.79, 6.32, 6.84, 7.37, 7.89, 8.42, 8.95, 9.47, 10.0])
calories = np.array([7.99, 9.17, 11.91, 12.89, 9.70, 11.32, 15.49, 17.53, 20.27, 20.92, 24.76, 24.80, 26.36, 26.21, 29.41, 33.71, 31.72, 36.96, 32.63, 35.60])

# Case 2
anxiety = np.array([-4.0, -3.58, -3.16, -2.74, -2.32, -1.89, -1.47, -1.05, -0.63, -0.21, 0.21, 0.63, 1.05, 1.47, 1.89, 2.32, 2.74, 3.16, 3.58, 4.0])
self_esteem = np.array([41.02, 43.27, 39.14, 25.96, 23.08, 16.41, 14.31, 12.26, 1.91, 5.65, 3.80, -5.23, -10.89, -2.85, -3.41, -3.34, -0.75, 13.14, 11.03, 17.77])
```

### Scatter Plots for both cases

```python
plt.scatter(exercise, calories, alpha=0.5)
plt.title('Exercise Hours vs. Calories burnt')
plt.xlabel('Exercise Hours')
plt.ylabel('Calories burnt')
plt.grid(True)
plt.show()
```

##### Looks linear probably Linear Regression would work well.

```python
plt.scatter(anxiety, self_esteem, alpha=0.5)
plt.title('Anxity Level vs. Self Esteem')
plt.xlabel('Anxity Level')
plt.ylabel('Self Esteem')
plt.grid(True)
plt.show()
```

##### Does not look linear probably Linear Regression would not work at all. Looks abit like a parabula.

### Linear Reggression for case #1

```python
res = np.polyfit(exercise, calories, deg=1)
equation = f"y = {res[0]:.3f}x + {res[1]:.3f}"

model = LinearRegression()
model.fit(exercise.reshape(-1, 1), calories)
prediction = model.predict(exercise.reshape(-1, 1))

plt.scatter(exercise.reshape(-1, 1)  , calories, color='blue', label='Data points', alpha=0.5)
plt.plot(exercise.reshape(-1, 1)  , prediction, color='red', label='Regression line')

plt.title('Linear Regression - Exercise Hours vs. Calories burnt')
plt.xlabel('Exercise Hours')
plt.ylabel('Calories burnt')
plt.grid(True)
plt.legend()
plt.text(5, 18, equation, fontsize=12)
plt.show()
```

```python
mse = mean_squared_error(calories, prediction)
print(f'Mean Squared Error: {mse}')

rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")

r2 = model.score(exercise.reshape(-1, 1), calories)
print(f'R^2: {r2}')
```

#### Does the model fit? 

I think it does.

### Polynomial Reggression for case #2

```python
coefficients = np.polyfit(anxiety, self_esteem, deg=2)
equation = f"y = {coefficients[2]:.2f} + ({coefficients[1]:.2f})x + ({coefficients[0]:.2f})x²"
print(f"Polynomial equation(polyfit): {equation}")
```

```python
poly_function = np.poly1d(coefficients)
prediction = poly_function(anxiety.reshape(-1, 1))

x_curve = np.linspace(-5, 5, 100)
y_curve = poly_function(x_curve)

plt.figure(figsize=(10, 6))
plt.scatter(anxiety, self_esteem, color='blue', label='Data points')
plt.plot(x_curve, y_curve, color='red', label='Polynomial regression')

plt.title('Polynomial Regression - Anxity Level vs. Self Esteem')
plt.xlabel('Anxity Level')
plt.ylabel('Self Esteem')
plt.grid(True)
plt.legend()

# Display equation on the graph
plt.text(-0.5, 10, equation, fontsize=12)
plt.show()
```

Formula for R²:

R^2=1−(∑(yi−y^i)^2)/(∑(yi−yˉ)^2)


```python
mse = mean_squared_error(self_esteem, prediction)
print(f'Mean Squared Error: {mse}')

rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")

numirator = np.sum((self_esteem - np.mean(self_esteem))**2)
denominator = np.sum((self_esteem - prediction)**2)
r2 = 1 - (numirator / denominator)
print(f'R^2: {r2}')
```

Sort of... the r^2 is realy good but the rmse... i'd like it to be much lower. It's not optimal.
